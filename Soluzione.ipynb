## 7. For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text.

```python
#First let's retrieve a clean list of all authors
clean_df = goodbooks['authors'].str.split(',', expand = True)
clean_df["shared_review_text"] = goodbooks['work_text_reviews_count'].values
clean_df.tail() 
```

```python
#We divide authors by chunks

a1 = clean_df[[0,'shared_review_text']]
a1.dropna(inplace=True)

a2 = clean_df[[1,'shared_review_text']]
a2.dropna(inplace=True)

a3 = clean_df[[2,'shared_review_text']]
a3.dropna(inplace=True)

a4 = clean_df[[3,'shared_review_text']]
a4.dropna(inplace=True)

a5 = clean_df[[4,'shared_review_text']]
a5.dropna(inplace=True)

#renaming chunks for an easier concatenation
a2.rename(columns={1: 0},inplace=True)
a3.rename(columns={2: 0},inplace=True)
a4.rename(columns={3: 0},inplace=True)
a5.rename(columns={4: 0},inplace=True)
```

```python
#We concatanete the chunks and summarize the results
authorsharedreview = pd.concat([a1,a2,a3,a4,a5])
authorsharedreview.rename(columns={0:"author"},inplace=True)
authorsharedreview.groupby('author')['shared_review_text'].sum()
print(authorsharedreview)
```
